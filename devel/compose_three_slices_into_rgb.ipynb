```python
from pathlib import Path
import skimage.io
from matplotlib import pyplot as plt

pth = Path("F:\Dataset\PNG\PNG")
allfiles = list(pth.glob("*.jpg"))
#allfiles = allfiles[:5]
```

```python

for i in range(0, len(allfiles)):
    name=str(allfiles[i])
    #name=name.replace("F:\Dataset\PNG\PNG\T" , "")#r
    #name="T"+name
    name1=Path("F:\Dataset\PNG\Png3in1")
    name=Path(name).stem+".jpg"
    im1 = skimage.io.imread(allfiles[i])  # Red
    if i > 0:
        im0 = skimage.io.imread(allfiles[i-1]) # Green
        im1[:,:,0] = im0[:,:,0]
    if (i + 1) < len(allfiles):
        im2 = skimage.io.imread(allfiles[i+1]) # Blue
        im1[:,:,2] = im2[:,:,2]
        
    #plt.figure(figsize=(10,10))
    #plt.imshow(im1)
    #plt.savefig(name)    # uklada s popiskama
    im1.shape
    skimage.io.imsave(name1/name,im1)
```
